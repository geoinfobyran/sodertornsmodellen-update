# Bostäder

Greger har slagit ihop tidigare data (2012-2016) med data från 2017 och räknat om andelar så att de blir rätt.

```python
df = pd.read_csv('./data/Bostader_2012-2017.csv', sep=';', decimal=',')
```

```python
df.dtypes
```

```python
df[df['område'] == "1220130 Östermalmstorg"]
```

```python
df = df.rename(columns={
    "tid": "year",
    "område": "basomrade",
    "Lght_Andel_Hyresrätter": "andel_hyres",
    "Lght_Andel_Bostadsrätter": "andel_bostadsr",
    "Lght_Andel_Äganderätter": "andel_aegender"
})
```

```python
# Format basomrade column
df['basomrade'] = df['basomrade'].str.lower()
df['basomrade'] = df['basomrade'].str.replace(' ', '_')
df['basomrade'] = df['basomrade'].str.replace('å', 'a')
df['basomrade'] = df['basomrade'].str.replace('ä', 'a')
df['basomrade'] = df['basomrade'].str.replace('ö', 'o')
```

```python
df[df['basomrade'] == "3710540_vastra_skogen"]
```

## Rensa bort 0:or för områden utan bostäder

Om `Lght total` är noll så sätter vi andelarna till `NaN`

```python
df.loc[df['Lght total'] == 0, ['andel_hyres','andel_bostadsr', 'andel_aegender']] = np.nan
```

```python
df
```

## Exportera resp. indikator

### Andel hyresrätter

```python
andel_hyres = df[['basomrade','year','andel_hyres']]
# andel_hyres
```

```python
andel_hyres.to_csv('./data/ddf--datapoints--andel_hyres--by--basomrade--year.csv', index=False)
```

### Andel bostadsrätter

```python
andel_bostadsr = df[['basomrade','year','andel_bostadsr']]
# andel_bostadsr
```

```python
andel_bostadsr.to_csv('./data/ddf--datapoints--andel_bostadsr--by--basomrade--year.csv', index=False)
```

### Andel äganderätter

```python
andel_aegender = df[['basomrade','year','andel_aegender']]
# andel_aegender
```

```python
andel_aegender.to_csv('./data/ddf--datapoints--andel_aegender--by--basomrade--year.csv', index=False)
```
