```python
import os
import glob

from functools import reduce

import numpy as np
import pandas as pd

%matplotlib inline
```

```python
files = glob.glob('../ddf--sodertornsmodellen-new/ddf--sodertornsmodellen--src/ddf--datapoints--*--by--basomrade--year.csv')
```

```python
files
```

```python
def readFile(file):
    df = pd.read_csv(file)
    df['year'] = pd.to_datetime(df['year'].astype('str'))
    return df
```

```python
dflist = list(map(readFile, files))
```

```python
df = reduce(lambda x, y: pd.merge(x, y, how='left'), dflist)
```
