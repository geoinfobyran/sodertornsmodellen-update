```python
import os
import glob

from functools import reduce

import numpy as np
import pandas as pd

%matplotlib inline
```

```python
files = glob.glob('../ddf--sodertornsmodellen-new/ddf--sodertornsmodellen--src/ddf--datapoints--*--by--basomrade--year.csv')
```

```python
files
```

```python
def readFile(file):
    df = pd.read_csv(file)
#     df['year'] = pd.to_datetime(df['year'].astype('str'))
    df['concept'] = file.split('/')[-1].split('--')[2]
    df = df.groupby('concept').agg({'year': ['min', 'max']})
    return df
```

```python
dflist = list(map(readFile, files))
```

```python
yearlist = pd.concat(dflist)
```

```python
yearlist = yearlist.reset_index()
```

```python
concepts = pd.read_csv('../ddf--sodertornsmodellen-new/ddf--sodertornsmodellen--src/ddf--concepts.csv')[['concept', 'name']]
```

```python
comb = pd.merge(yearlist, concepts, on='concept')

comb[['concept', 'name', 'year']]
# comb['yearlist']
# yearlist.shape
```

```python
yearlist.columns = yearlist.columns.get_level_values(0)
```
